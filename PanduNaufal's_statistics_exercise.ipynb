```python
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Membuat DataFrame
data = {
    "Student_ID": list(range(1, 16)),
    "Age": [18, 19, 18, 20, 21, 22, 19, 20, 18, 21, 19, 20, 22, 21, 18],
    "Gender": ["Male", "Female", "Female", "Male", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male"],
    "Hours_Studied": [5, 8, 6, 7, 10, 4, 9, 5, 6, 8, 7, 5, 4, 9, 6],
    "Math_Score": [78, 85, 72, 80, 95, 60, 88, 75, 70, 90, 85, 74, 65, 92, 80],
    "English_Score": [82, 88, 78, 84, 92, 65, 86, 79, 72, 91, 87, 76, 68, 93, 83],
    "Passed_Math": [1, 1, 1, 1, 1, 0, 1, 1, 1, 1, 1, 1, 0, 1, 1],
    "Passed_English": [1, 1, 1, 1, 1, 0, 1, 1, 1, 1, 1, 1, 0, 1, 1],
    "Attendance_Rate": [90, 95, 85, 80, 100, 70, 88, 84, 92, 96, 89, 86, 75, 97, 91],
    "Class_Section": ["Class A", "Class B", "Class C", "Class A", "Class B", "Class C",
                      "Class A", "Class B", "Class C", "Class A", "Class B", "Class C",
                      "Class A", "Class B", "Class C"]
}
df = pd.DataFrame(data)

# 1. ANALISIS SKOR MATEMATIKA
math_scores = df['Math_Score']

# Statistik deskriptif
mean = math_scores.mean()
q1, q2, q3 = math_scores.quantile([0.25, 0.5, 0.75])
p1, p5, p95, p99 = np.percentile(math_scores, [1, 5, 95, 99])
skewness = stats.skew(math_scores)
kurtosis = stats.kurtosis(math_scores)

# Cetak hasil analisis deskriptif
print("Analisis Statistik Deskriptif Skor Matematika:")
print(f"Rata-rata: {mean}")
print(f"Q1: {q1}, Median (Q2): {q2}, Q3: {q3}")
print(f"Persentil: P1={p1}, P5={p5}, P95={p95}, P99={p99}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}\n")

# 2. INTERVAL KEPERCAYAAN UNTUK SKOR MATEMATIKA
# CI untuk rata-rata (varian diketahui = 4)
var_known = 4
se_known = np.sqrt(var_known / len(math_scores))
z_critical = stats.norm.ppf(0.975)  # Z nilai kritis untuk alpha=5%
ci_mean_known = (mean - z_critical * se_known, mean + z_critical * se_known)

# CI untuk rata-rata (varian tidak diketahui)
se_unknown = stats.sem(math_scores)
t_critical = stats.t.ppf(0.975, len(math_scores) - 1)
ci_mean_unknown = (mean - t_critical * se_unknown, mean + t_critical * se_unknown)

# CI untuk variansi
chi2_lower = stats.chi2.ppf(0.025, len(math_scores) - 1)
chi2_upper = stats.chi2.ppf(0.975, len(math_scores) - 1)
ci_variance = ((len(math_scores) - 1) * math_scores.var() / chi2_upper,
               (len(math_scores) - 1) * math_scores.var() / chi2_lower)

# CI untuk proporsi skor tinggi (>85)
high_scores = (math_scores > 85).sum()
p_high = high_scores / len(math_scores)
se_prop = np.sqrt(p_high * (1 - p_high) / len(math_scores))
ci_prop = (p_high - z_critical * se_prop, p_high + z_critical * se_prop)

# Cetak hasil interval kepercayaan
print("Interval Kepercayaan:")
print(f"Rata-rata (varian diketahui): {ci_mean_known}")
print(f"Rata-rata (varian tidak diketahui): {ci_mean_unknown}")
print(f"Variansi: {ci_variance}")
print(f"Proporsi skor tinggi (>85): {ci_prop}\n")

# 3. INTERVAL KEPERCAYAAN UNTUK UMUR (LULUS VS TIDAK LULUS MATEMATIKA)
ages_passed = df[df['Passed_Math'] == 1]['Age']
ages_failed = df[df['Passed_Math'] == 0]['Age']

# CI untuk perbedaan rata-rata (varian diketahui)
mean_diff = ages_passed.mean() - ages_failed.mean()
var_passed = 5
var_failed = 4
se_diff = np.sqrt(var_passed / len(ages_passed) + var_failed / len(ages_failed))
ci_mean_diff_known = (mean_diff - z_critical * se_diff, mean_diff + z_critical * se_diff)

# CI untuk perbedaan rata-rata (varian tidak diketahui)
se_diff_unknown = np.sqrt(ages_passed.var(ddof=1) / len(ages_passed) + ages_failed.var(ddof=1) / len(ages_failed))
t_critical_diff = stats.t.ppf(0.975, min(len(ages_passed), len(ages_failed)) - 1)
ci_mean_diff_unknown = (mean_diff - t_critical_diff * se_diff_unknown, mean_diff + t_critical_diff * se_diff_unknown)

# CI untuk rasio variansi
if ages_failed.var() == 0:
    print("Variansi kelompok tidak lulus matematika adalah nol. Tidak dapat menghitung rasio variansi.")
    ci_variance_ratio = None
else:
    f_lower = stats.f.ppf(0.025, len(ages_passed) - 1, len(ages_failed) - 1)
    f_upper = stats.f.ppf(0.975, len(ages_passed) - 1, len(ages_failed) - 1)
    var_ratio = ages_passed.var() / ages_failed.var()
    ci_variance_ratio = (var_ratio / f_upper, var_ratio / f_lower)

# Cetak hasil CI rasio variansi
if ci_variance_ratio is not None:
    print(f"Rasio variansi: {ci_variance_ratio}\n")
else:
    print("Tidak dapat menghitung rasio variansi karena variansi nol.\n")


# Cetak hasil CI
print("Interval Kepercayaan untuk Umur (Lulus vs Tidak Lulus):")
print(f"Perbedaan rata-rata (varian diketahui): {ci_mean_diff_known}")
print(f"Perbedaan rata-rata (varian tidak diketahui): {ci_mean_diff_unknown}")
print(f"Rasio variansi: {ci_variance_ratio}\n")

# 9. UJI ANOVA UNTUK NILAI BAHASA INGGRIS
anova_model = ols('English_Score ~ Class_Section', data=df).fit()
anova_results = anova_lm(anova_model)

# Bandingkan nilai p-value dengan alpha=0.05
alpha = 0.05
anova_p_value = anova_results['PR(>F)'][0]

print("Hasil Uji ANOVA (Nilai Bahasa Inggris Berdasarkan Kelas):")
print(anova_results)
if anova_p_value < alpha:
    print("Keputusan: Tolak H0. Terdapat perbedaan signifikan rata-rata nilai Bahasa Inggris antar kelas.")
else:
    print("Keputusan: Gagal menolak H0. Tidak ada cukup bukti untuk menyimpulkan perbedaan rata-rata nilai Bahasa Inggris antar kelas.")
```
